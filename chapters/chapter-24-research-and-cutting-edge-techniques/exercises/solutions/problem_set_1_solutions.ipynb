### Task 1: Softmax is a distribution

Assert that softmax over a random matrix produces rows that sum to 1 and are non-negative.

```python
import sys; sys.path.insert(0, '../../scripts')
import numpy as np
from attention import softmax
def check_softmax():
    p = softmax(np.random.default_rng(0).standard_normal((4, 5)), axis=1)
    assert np.allclose(p.sum(axis=1), 1.0)
    assert (p >= 0).all()
    return p.shape
print(check_softmax())
```

### Task 2: Attention output shape

Assert attention output has shape (n_q, d_v) and weights sum to 1 per query.

```python
import sys; sys.path.insert(0, '../../scripts')
import numpy as np
from attention import scaled_dot_product_attention
def check_attention():
    rng = np.random.default_rng(1)
    out, attn = scaled_dot_product_attention(rng.standard_normal((2, 8)),
                                             rng.standard_normal((6, 8)),
                                             rng.standard_normal((6, 3)))
    assert out.shape == (2, 3)
    assert np.allclose(attn.sum(axis=1), 1.0)
    return out.shape
print(check_attention())
```
