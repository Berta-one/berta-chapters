### Task 1: Top-k selects k experts

Assert the router returns exactly k experts per token and weights that sum to 1.

```python
import sys; sys.path.insert(0, '../../scripts')
import numpy as np
from moe import top_k_gating
def check_gating():
    rng = np.random.default_rng(2)
    idx, w = top_k_gating(rng.standard_normal((5, 16)), rng.standard_normal((16, 4)), k=2)
    assert idx.shape == (5, 2)
    assert np.allclose(w.sum(axis=1), 1.0)
    return idx.shape
print(check_gating())
```

### Task 2: MoE forward shape

Run the sparse forward pass and assert the output shape matches the experts' output dimension.

```python
import sys; sys.path.insert(0, '../../scripts')
import numpy as np
from moe import moe_forward
def check_moe():
    rng = np.random.default_rng(3)
    x = rng.standard_normal((4, 16))
    W = rng.standard_normal((16, 4))
    experts = [lambda x, s=s: x * s for s in (1.0, 2.0, 3.0, 4.0)]
    y, idx, w = moe_forward(x, W, experts, k=2)
    assert y.shape == (4, 16)
    return y.shape
print(check_moe())
```
