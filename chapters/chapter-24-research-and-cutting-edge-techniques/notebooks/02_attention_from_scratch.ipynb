### Key concepts

- **Attention** — a softmax-weighted average of values, keyed by query-key similarity.
- **Scaling** — dividing by sqrt(d_k) keeps softmax gradients healthy.
- **Causal mask** — prevents a position from attending to the future.
- **Invariant** — every attention row is a probability distribution.

## 1-2. The core operation

`Attention(Q, K, V) = softmax(Q Kᵀ / sqrt(d_k)) V`. We feed in random Q, K, V and confirm the output has V's value dimension and the weights sum to one.

```python
import sys
sys.path.insert(0, "../scripts")
import numpy as np
from attention import scaled_dot_product_attention

rng = np.random.default_rng(24)
Q = rng.standard_normal((3, 8))
K = rng.standard_normal((5, 8))
V = rng.standard_normal((5, 4))
out, attn = scaled_dot_product_attention(Q, K, V)
print("output shape:", out.shape, "(n_q, d_v)")
print("attention rows sum to 1:", np.allclose(attn.sum(axis=1), 1.0))
```

## 3. Causal masking

For autoregressive models, a lower-triangular mask blocks attention to future positions.

```python
from attention import causal_mask
mask = causal_mask(5)
Qs = rng.standard_normal((5, 8))
Ks = rng.standard_normal((5, 8))
Vs = rng.standard_normal((5, 4))
_, attn_masked = scaled_dot_product_attention(Qs, Ks, Vs, mask=mask)
print("upper triangle is ~0:", np.allclose(np.triu(attn_masked, k=1), 0, atol=1e-6))
```
