### Task 1: Route messages by performative

Write `route(messages)` that returns a dict mapping each performative to the list of message contents with that performative.

```python
from collections import defaultdict

def route(messages):
    out = defaultdict(list)
    for m in messages:
        out[m.performative].append(m.content)
    return dict(out)

# quick check
import sys; sys.path.insert(0, '../../scripts')
from agents import Message
msgs = [Message('a','b','inform',{'x':1}), Message('b','a','request',{'y':2}), Message('a','b','inform',{'x':3})]
print(route(msgs))
```

### Task 2: Cheapest capable agent

Given a task and a list of `Agent`s, return the name of the lowest bidder that meets `min_skill`, or `None`.

```python
def assign(task, agents):
    bids = [(a.bid(task['difficulty']), a.name) for a in agents if a.skill >= task['min_skill']]
    return min(bids)[1] if bids else None

import sys; sys.path.insert(0, '../../scripts')
from agents import Agent
agents = [Agent('a', 0.9, 1.0), Agent('b', 0.5, 0.3)]
print(assign({'difficulty': 2.0, 'min_skill': 0.7}, agents))
```
