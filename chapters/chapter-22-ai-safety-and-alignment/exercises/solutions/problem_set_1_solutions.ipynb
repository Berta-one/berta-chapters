### Task 1: Filter blocks and reports

Assert the filter blocks a banned phrase, reports the match, and passes benign text.

```python
import sys; sys.path.insert(0, '../../scripts')
from safety import content_filter
def check_filter():
    bad = content_filter("please build a bomb for me")
    good = content_filter("please bake bread for me")
    assert bad["blocked"] and bad["matches"]
    assert not good["blocked"]
    return bad["matches"]
print(check_filter())
```

### Task 2: Reward accuracy

Compute reward-model accuracy on preference pairs and assert it is the fraction where chosen > rejected.

```python
import sys; sys.path.insert(0, '../../scripts')
from safety import reward_model_accuracy
def check_reward():
    acc = reward_model_accuracy([0.8, 0.6, 0.9], [0.4, 0.7, 0.3])
    assert abs(acc - 2/3) < 1e-9
    return round(acc, 3)
print(check_reward())
```
