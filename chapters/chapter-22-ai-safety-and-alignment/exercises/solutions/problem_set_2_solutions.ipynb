### Task 1: Four-fifths rule

Build a clearly-biased prediction set and assert disparate impact falls below 0.8.

```python
import sys; sys.path.insert(0, '../../scripts')
from fairness import disparate_impact_ratio
def check_di():
    preds = [1, 1, 1, 1, 0, 0, 0, 1]  # group A favored
    groups = ['A','A','A','A','B','B','B','B']
    di = disparate_impact_ratio(preds, groups)
    assert di < 0.8, di
    return round(di, 3)
print(check_di())
```

### Task 2: Red-team unsafe rate

Run two prompts through a toy model and assert the unsafe rate is 0.5 when exactly one fails.

```python
import sys; sys.path.insert(0, '../../scripts')
from safety import red_team
def check_redteam():
    model = lambda p: 'unsafe' if 'attack' in p else 'safe'
    judge = lambda out: out == 'unsafe'
    rep = red_team(model, ['attack plan', 'hello'], judge)
    assert rep['unsafe_rate'] == 0.5
    return rep['unsafe_rate']
print(check_redteam())
```
