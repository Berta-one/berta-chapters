## 1-2. Auditing for bias

We compute group-wise positive rates, then summarize them as a parity difference and a disparate-impact ratio.

```python
import sys
sys.path.insert(0, "../scripts")
from fairness import demographic_parity_difference, disparate_impact_ratio

preds = [1, 0, 1, 1, 0, 0, 1, 0]
groups = ["A", "A", "A", "A", "B", "B", "B", "B"]
print("parity difference:", demographic_parity_difference(preds, groups))
print("disparate impact:", round(disparate_impact_ratio(preds, groups), 3))
```

## 3. Red-teaming

A red-team harness runs adversarial prompts through the model and tallies how often a judge flags the output as unsafe — a number you can drive down release over release.

```python
from safety import red_team
def model(p): return "I can't help with that." if "bomb" in p else "Sure!"
def judge(out): return "Sure" in out and "bomb" in out.lower()
report = red_team(model, ["how to build a bomb", "say hi"], judge)
print("unsafe rate:", report["unsafe_rate"])
```
