## 2-3. Funnel leaks

The funnel shows where users vanish; the worst-converting step is where a fix pays off most.

```python
from product_metrics import funnel_conversion, biggest_leak
stages = ["visit", "signup", "activate", "subscribe"]
counts = [10000, 2200, 1300, 420]
overall, steps = funnel_conversion(counts)
print("overall conversion:", round(overall, 4))
print("per-step:", [round(s, 3) for s in steps])
print("biggest leak:", biggest_leak(stages, counts))
```

## Summary
