## 1-2. RICE scoring

RICE forces every feature onto the same scale: how many users, how much it helps, how sure we are, divided by the work.

## 3. Rank the backlog

Sorting by RICE turns a pile of opinions into a defensible order.

```python
from product_metrics import rank_backlog
backlog = [
    {"name": "ai_autocomplete", "reach": 8000, "impact": 2.0, "confidence": 0.8, "effort": 5},
    {"name": "dark_mode", "reach": 5000, "impact": 0.5, "confidence": 0.9, "effort": 1},
    {"name": "team_sharing", "reach": 3000, "impact": 1.0, "confidence": 0.7, "effort": 3},
]
for name, score in rank_backlog(backlog):
    print(f"{name:18s} {score:8.1f}")
```
