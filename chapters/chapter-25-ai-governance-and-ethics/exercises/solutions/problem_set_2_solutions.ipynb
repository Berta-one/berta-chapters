### Task 1: Card contains required sections

Generate a model card from complete metadata and assert every required section heading appears.

```python
import sys; sys.path.insert(0, '../../scripts')
from governance import generate_model_card
from config import REQUIRED_CARD_FIELDS
def check_card():
    meta = {f: f"value for {f}" for f in REQUIRED_CARD_FIELDS}
    card = generate_model_card(meta)
    for f in REQUIRED_CARD_FIELDS:
        assert f.replace("_", " ").title() in card
    return "complete"
print(check_card())
```
