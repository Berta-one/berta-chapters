### Task 1: Drawdown is non-positive

Build an equity curve from returns and assert max drawdown is at most 0.

```python
import sys; sys.path.insert(0, '../../scripts')
from metrics import equity_curve, max_drawdown
def check_drawdown():
    eq = equity_curve([0.05, -0.10, 0.03, -0.02])
    dd = max_drawdown(eq)
    assert dd <= 0.0
    return round(dd, 4)
print(check_drawdown())
```

### Task 2: Sharpe rewards consistency

Show that a steady positive return stream has a higher Sharpe than a volatile one with the same mean.

```python
import sys; sys.path.insert(0, '../../scripts')
from metrics import sharpe_ratio
def check_sharpe():
    steady = [0.012, 0.011, 0.013, 0.010]
    volatile = [0.05, -0.03, 0.04, -0.02]
    assert sharpe_ratio(steady) > sharpe_ratio(volatile)
    return round(sharpe_ratio(steady), 2), round(sharpe_ratio(volatile), 2)
print(check_sharpe())
```
