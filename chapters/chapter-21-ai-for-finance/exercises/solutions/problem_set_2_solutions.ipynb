### Task 1: Backtest returns a valid equity curve

Run the backtester and assert the equity curve is positive and the result dict has the expected keys.

```python
import sys; sys.path.insert(0, '../../scripts')
from backtest import ma_crossover_signal, backtest
def check_backtest():
    prices = [100, 101, 102, 101, 103, 105, 104, 106, 108, 107]
    sig = ma_crossover_signal(prices)
    res = backtest(prices, sig, cost_per_trade=0.0005)
    assert set(res) >= {'equity', 'sharpe', 'max_drawdown', 'total_return'}
    assert (res['equity'] > 0).all()
    return round(res['total_return'], 4)
print(check_backtest())
```

### Task 2: Costs reduce returns

Show that adding transaction costs never increases total return.

```python
import sys; sys.path.insert(0, '../../scripts')
from backtest import ma_crossover_signal, backtest
def check_costs():
    prices = [100, 102, 101, 104, 103, 106, 105, 108]
    sig = ma_crossover_signal(prices)
    free = backtest(prices, sig, 0.0)['total_return']
    costly = backtest(prices, sig, 0.001)['total_return']
    assert costly <= free + 1e-12
    return round(free - costly, 5)
print(check_costs())
```
