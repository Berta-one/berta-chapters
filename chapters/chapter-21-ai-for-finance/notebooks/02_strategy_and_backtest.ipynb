### Key concepts

- **Crossover** — fast MA above slow MA signals an uptrend (go long).
- **Look-ahead bias** — using information you wouldn't have had in real time.
- **Lagging** — act on bar t's signal at bar t+1, never the same bar.
- **Transaction costs** — every position change costs money and erodes turnover-heavy strategies.

## 1-2. Build the signal

The crossover produces a long/flat position; only past prices feed each signal value, so there is no peeking.

```python
import sys
sys.path.insert(0, "../scripts")
from backtest import ma_crossover_signal

prices = [100, 101.2, 100.5, 102.1, 103.4, 102.8, 104.6, 105.9, 105.1, 106.7]
sig = ma_crossover_signal(prices, fast=3, slow=5)
print("signal:", list(sig))
```

## 3. Backtest with costs

The backtester lags the signal by one bar and subtracts a cost on every position change.

```python
from backtest import backtest
res = backtest(prices, sig, cost_per_trade=0.0005)
print("total return:", round(res["total_return"], 4))
print("Sharpe:", round(res["sharpe"], 3))
print("max drawdown:", round(res["max_drawdown"], 4))
```

## Summary

A crossover signal plus a lagged, cost-aware backtester gives an honest first estimate of a strategy. The lag is what keeps the result from being a fantasy.
