### Key concepts

- **Look-ahead bias** — the single most common backtest error.
- **Survivorship bias** — testing only on assets that still exist flatters results.
- **Multiple testing** — try enough strategies and one looks great by luck.
- **Signal-to-noise** — financial returns are mostly noise; humility is a strategy.

## 1. Detecting look-ahead

A quick check: a correct backtest must give different (worse) results than one that cheats by using the *same-bar* signal.

```python
import sys
sys.path.insert(0, "../scripts")
import numpy as np
from metrics import simple_returns
from backtest import ma_crossover_signal

prices = [100, 101.2, 100.5, 102.1, 103.4, 102.8, 104.6, 105.9, 105.1, 106.7]
sig = ma_crossover_signal(prices)
rets = simple_returns(prices)
honest = (rets * np.asarray(sig, float)[:-1]).sum()   # lagged
cheating = (rets * np.asarray(sig, float)[1:]).sum()   # same-bar (look-ahead)
print("honest:", round(honest, 4), "| cheating:", round(cheating, 4))
```
