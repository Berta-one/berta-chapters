### Key concepts

- **Simple return** — `p[t]/p[t-1] - 1`; aggregates across assets.
- **Log return** — additive over time; convenient for statistics.
- **Sharpe ratio** — excess return per unit of volatility, annualized.
- **Maximum drawdown** — the worst peak-to-trough loss; what investors feel.

## 1-2. Returns and volatility

We load a price series and compute both return conventions, then annualize the volatility.

```python
import sys
sys.path.insert(0, "../scripts")
import numpy as np
from metrics import simple_returns, log_returns, volatility

prices = [100, 101.2, 100.5, 102.1, 103.4, 102.8, 104.6, 105.9, 105.1, 106.7]
r = simple_returns(prices)
print("simple returns:", np.round(r, 4))
print("annualized vol:", round(volatility(r), 4))
```

## 3-4. Risk-adjusted performance

The Sharpe ratio rewards return per unit of risk; maximum drawdown captures the worst loss along the way.

```python
from metrics import sharpe_ratio, equity_curve, max_drawdown
print("Sharpe:", round(sharpe_ratio(r), 3))
eq = equity_curve(r)
print("max drawdown:", round(max_drawdown(eq), 4))
```
