### Task 1: Discounted return

Given a list of rewards and gamma, compute the discounted return from t=0.

```python
def discounted_return(rewards, gamma):
    return sum((gamma ** t) * r for t, r in enumerate(rewards))
print(round(discounted_return([0, 0, 1], 0.9), 4))
```
