### Task 1: Train and reach the goal

Train Q-learning on the gridworld and assert the greedy policy reaches the goal within the step budget.

```python
import sys; sys.path.insert(0, '../../scripts')
from gridworld import GridWorld
from q_learning import train, greedy_policy, run_policy
from config import LAYOUT
def trained_reaches_goal():
    env = GridWorld(LAYOUT)
    Q, _ = train(env, episodes=800)
    steps = run_policy(env, greedy_policy(Q))
    assert steps < 100
    return steps
print(trained_reaches_goal())
```

### Task 2: Epsilon controls exploration

Explain in a comment, then show, that epsilon=0 from an all-zero Q-table gets stuck while epsilon>0 explores.

```python
import sys; sys.path.insert(0, '../../scripts')
from gridworld import GridWorld
from q_learning import train, greedy_policy, run_policy
from config import LAYOUT
def epsilon_matters():
    env = GridWorld(LAYOUT)
    Q, _ = train(env, episodes=800, epsilon=0.2)
    return run_policy(env, greedy_policy(Q))
print(epsilon_matters())
```
