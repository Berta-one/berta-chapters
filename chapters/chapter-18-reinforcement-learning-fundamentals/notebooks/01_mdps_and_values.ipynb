### Key concepts

- **MDP** — the formal model of sequential decisions under uncertainty.
- **Discount factor (gamma)** — how much future reward is worth now.
- **Value function** — expected return from a state (or state-action pair).
- **Bellman equation** — value defined recursively in terms of successor values.

## 1. Build a gridworld MDP

The agent moves on a grid; walls block movement, the goal ends the episode with +1, and each step costs a little to encourage short paths.

```python
import sys
sys.path.insert(0, "../scripts")
from gridworld import GridWorld
from config import LAYOUT

env = GridWorld(LAYOUT)
print("start:", env.start, "goal:", env.goal)
print("num states:", len(env.states()))
```

## 4. Value iteration

With a known model we can compute optimal values directly by repeatedly applying the Bellman optimality backup.

```python
V = env.value_iteration(gamma=0.95)
print("V(start) =", round(V[env.start], 3))
print("V(goal)  =", round(V[env.goal], 3))
```
