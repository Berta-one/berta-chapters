## 2. REINFORCE on a bandit

A minimal policy gradient: parameterize action probabilities with a softmax over preferences and nudge them toward higher-reward actions.

```python
import numpy as np
rng = np.random.default_rng(18)
true_reward = np.array([0.2, 0.5, 0.9])  # 3 arms
prefs = np.zeros(3)
alpha = 0.1
for t in range(2000):
    p = np.exp(prefs) / np.exp(prefs).sum()
    a = rng.choice(3, p=p)
    r = true_reward[a] + 0.01 * rng.standard_normal()
    grad = -p
    grad[a] += 1.0
    prefs += alpha * r * grad
print("final action probs:", np.round(np.exp(prefs)/np.exp(prefs).sum(), 3))
```
