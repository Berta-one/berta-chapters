## 1-2. Canary routing

Hashing the user key gives a stable, uniform split: roughly the target fraction of users hit the canary, and each user's assignment is consistent.

```python
import sys
sys.path.insert(0, "../scripts")
from reliability import canary_route

assignments = [canary_route(f"user-{i}", 0.1) for i in range(2000)]
frac = assignments.count("canary") / len(assignments)
print("canary fraction (~0.10):", round(frac, 3))
print("user-7 is stable across calls:",
      canary_route("user-7", 0.1) == canary_route("user-7", 0.1))
```
