### Task 1: Canary split is stable and ~correct

Assert the canary fraction over 5000 users is within 0.02 of the target and each user's assignment is stable.

```python
import sys; sys.path.insert(0, '../../scripts')
from reliability import canary_route
def check_canary():
    n = 5000
    frac = sum(canary_route(i, 0.2) == 'canary' for i in range(n)) / n
    assert abs(frac - 0.2) < 0.02, frac
    assert canary_route(42, 0.2) == canary_route(42, 0.2)
    return round(frac, 3)
print(check_canary())
```

### Task 2: p99 latency

Compute the p99 latency of a list and confirm it ignores the bulk and captures the tail.

```python
import sys; sys.path.insert(0, '../../scripts')
from observability import percentile
def check_p99():
    vals = [10] * 99 + [1000]
    assert percentile(vals, 99) == 10
    assert percentile(vals, 100) == 1000
    return percentile(vals, 99)
print(check_p99())
```
