### Task 1: Bucket allows exactly the burst

Create a bucket with capacity 3 and assert exactly 3 of 5 simultaneous requests are allowed.

```python
import sys; sys.path.insert(0, '../../scripts')
from reliability import TokenBucket
def check_bucket():
    tb = TokenBucket(rate=1.0, capacity=3.0)
    allowed = sum(tb.allow(now=0.0) for _ in range(5))
    assert allowed == 3, allowed
    return allowed
print(check_bucket())
```

### Task 2: Breaker opens after threshold

After `fail_threshold` failures, assert the breaker is open and fails fast.

```python
import sys; sys.path.insert(0, '../../scripts')
from reliability import CircuitBreaker, CircuitOpen
def check_breaker():
    cb = CircuitBreaker(fail_threshold=2, reset_timeout=10.0)
    def boom(): raise ValueError()
    for t in range(2):
        try: cb.call(boom, now=float(t))
        except ValueError: pass
    assert cb.state == 'open'
    try:
        cb.call(boom, now=2.0); return 'no-raise'
    except CircuitOpen:
        return 'fast-fail'
print(check_breaker())
```
