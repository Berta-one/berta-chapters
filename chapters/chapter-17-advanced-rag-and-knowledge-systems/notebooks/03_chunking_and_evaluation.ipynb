## 1. Chunking with overlap

A simple sliding window with overlap keeps context across boundaries.

```python
def chunk(text, size=8, overlap=3):
    words = text.split()
    step = max(size - overlap, 1)
    return [" ".join(words[i:i+size]) for i in range(0, len(words), step) if words[i:i+size]]

doc = "backpropagation computes gradients via the chain rule across every layer of the network"
for c in chunk(doc):
    print("-", c)
```
