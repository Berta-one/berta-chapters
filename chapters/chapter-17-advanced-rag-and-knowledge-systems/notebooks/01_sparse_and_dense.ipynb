### Key concepts

- **Dense retrieval** — embeddings capture meaning, catching paraphrases BM25 misses.
- **IDF** — rare terms are more informative than common ones.
- **Vocabulary mismatch** — the core problem dense retrieval solves.

## 1-2. BM25 from scratch

BM25 scores a document by summing, over query terms, an IDF weight times a saturating term-frequency factor with length normalization.

```python
import sys
sys.path.insert(0, "../scripts")
from hybrid_search import BM25

corpus = [
    "training neural networks with gradient descent",
    "the history of the roman empire",
    "backpropagation computes gradients via the chain rule",
]
bm25 = BM25(corpus)
print("ranking:", bm25.rank("how do gradients flow in backprop"))
```

## 3. Dense retrieval

With embeddings, a query and a document are vectors; cosine similarity measures their semantic closeness — even with no shared words.

```python
import numpy as np
from hybrid_search import dense_rank

# toy 3-dim embeddings (in practice from a sentence encoder)
doc_vecs = [[0.9, 0.1, 0.0], [0.0, 0.2, 0.9], [0.85, 0.15, 0.05]]
query_vec = [0.88, 0.12, 0.02]
print("dense ranking:", dense_rank(query_vec, doc_vecs))
```
