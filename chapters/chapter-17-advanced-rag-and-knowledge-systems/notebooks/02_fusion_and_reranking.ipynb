## 1-2. Reciprocal Rank Fusion

RRF adds `1/(k + rank)` from each ranking. Documents ranked highly by *either* retriever rise to the top, with no score calibration.

```python
import sys
sys.path.insert(0, "../scripts")
from hybrid_search import reciprocal_rank_fusion

sparse = [2, 0, 1]   # doc ids, best first
dense = [0, 2, 1]
print("fused:", reciprocal_rank_fusion([sparse, dense]))
```

## 3. Reranking

After fusion, rerank the top candidates with a more expensive, phrase-aware scorer to break ties precisely.

```python
from hybrid_search import rerank

corpus = [
    "training neural networks with gradient descent",
    "the history of the roman empire",
    "backpropagation computes gradients via the chain rule",
]
print("reranked:", rerank("chain rule", [0, 1, 2], corpus))
```
