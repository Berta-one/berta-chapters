### Task 1: Compute IDF

Given document frequency `n` and corpus size `N`, return the BM25 IDF `log(1 + (N - n + 0.5)/(n + 0.5))`.

```python
import math
def idf(n, N):
    return math.log(1 + (N - n + 0.5) / (n + 0.5))
print(round(idf(1, 5), 3), round(idf(4, 5), 3))
```
