## 1. Size across precisions

The same 100M-parameter model is 400 MB in float32 but only 100 MB in int8 — often the difference between fitting on a device or not.

```python
import sys
sys.path.insert(0, "../scripts")
from optimize import model_size_bytes

n = 100_000_000
for prec in ("float32", "float16", "int8"):
    mb = model_size_bytes(n, prec) / 1e6
    print(f"{prec:8s}: {mb:6.1f} MB")
```

## 2. Latency vs throughput

Larger batches amortize fixed overhead, raising throughput — but each individual request waits longer.

```python
from inference import batch_latency, throughput

for b in (1, 8, 32, 128):
    print(f"batch={b:4d}  latency={batch_latency(b):6.1f}ms  throughput={throughput(b):7.1f}/s")
```
