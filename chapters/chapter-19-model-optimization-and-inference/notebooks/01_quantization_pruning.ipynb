### Key concepts

- **Quantization** — map float weights to a small integer range via a scale factor.
- **Scale** — `max|W| / 127` for symmetric int8; sets the quantization step.
- **Pruning** — zeroing small weights; sparse models compress and can run faster.
- **Round-trip error** — bounded by half the quantization step.

## 1-2. int8 quantization

We map weights to the int8 range with a single scale factor and measure how much information we lose on the round trip.

```python
import sys
sys.path.insert(0, "../scripts")
import numpy as np
from optimize import quantize_int8, dequantize_int8, quantization_error

rng = np.random.default_rng(19)
W = rng.standard_normal((4, 4))
q, scale = quantize_int8(W)
print("scale:", round(scale, 4))
print("max round-trip error:", round(quantization_error(W), 5))
print("bound (scale/2):", round(scale / 2, 5))
```

## 3-4. Magnitude pruning

Removing the smallest-magnitude weights trades a little accuracy for sparsity that compresses and accelerates inference.

```python
from optimize import magnitude_prune, measure_sparsity

Wp = magnitude_prune(W, sparsity=0.5)
print("requested sparsity: 0.50")
print("actual sparsity:   ", measure_sparsity(Wp))
```

Quantization shrinks weights to int8 with error bounded by half the scale step; magnitude pruning removes the least important weights to a chosen sparsity. Both are nearly free accuracy-wise up to a point.
