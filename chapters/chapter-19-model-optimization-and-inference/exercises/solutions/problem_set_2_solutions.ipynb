### Task 1: Soft targets sum to one

Produce distillation targets at T=3 and assert they form a valid probability distribution that is softer than T=1.

```python
import sys; sys.path.insert(0, '../../scripts')
import numpy as np
from optimize import softmax_with_temperature
def check_soft_targets():
    logits = np.array([5.0, 1.0, 0.0])
    hot = softmax_with_temperature(logits, 1.0)
    soft = softmax_with_temperature(logits, 3.0)
    assert abs(soft.sum() - 1.0) < 1e-9
    assert soft.max() < hot.max()  # softer
    return np.round(soft, 3)
print(check_soft_targets())
```

### Task 2: int8 is 4x smaller

Assert that an int8 model is one quarter the size of the float32 version.

```python
import sys; sys.path.insert(0, '../../scripts')
from optimize import model_size_bytes
def check_size():
    n = 1_000_000
    assert model_size_bytes(n, 'float32') == 4 * model_size_bytes(n, 'int8')
    return model_size_bytes(n, 'int8')
print(check_size())
```
