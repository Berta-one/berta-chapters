### Task 1: Error is bounded by scale/2

Quantize a random matrix and assert the max round-trip error is at most scale/2 plus a tiny epsilon.

```python
import sys; sys.path.insert(0, '../../scripts')
import numpy as np
from optimize import quantize_int8, quantization_error
def check_quant_bound():
    rng = np.random.default_rng(0)
    W = rng.standard_normal((8, 8))
    _, scale = quantize_int8(W)
    err = quantization_error(W)
    assert err <= scale / 2 + 1e-9
    return round(err, 5)
print(check_quant_bound())
```

### Task 2: Prune to target sparsity

Prune a matrix to 70% sparsity and assert the measured sparsity matches within one element.

```python
import sys; sys.path.insert(0, '../../scripts')
import numpy as np
from optimize import magnitude_prune, measure_sparsity
def check_prune():
    rng = np.random.default_rng(1)
    W = rng.standard_normal((10, 10))
    Wp = magnitude_prune(W, 0.7)
    assert abs(measure_sparsity(Wp) - 0.7) <= 0.01
    return measure_sparsity(Wp)
print(check_prune())
```
